```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
input_img = cv2.imread("yuta.png")
#input_img = cv2.cvtColor(input_img,cv2.COLOR_BGR2RGB)

cv2.imshow("Image",input_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.axis('off')
plt.imshow(input_img)
plt.show()
```

```python
rows,cols,dim=input_img.shape
```

# Translation

```python
M=np.float32([[1,0,200],
             [0,1,250],
             [0,0,1]])

translated_img=cv2.warpPerspective(input_img,M,(cols,rows))

plt.axis('off')
plt.imshow(translated_img)
plt.show()
```

# Scaling

```python
M=np.float32([[1.5,0,0],
             [0,1.5,0],
             [0,0,1]])

scaled_img=cv2.warpPerspective(input_img,M,(cols,rows))

plt.axis('off')
plt.imshow(scaled_img)
plt.show()
```

# Image shearing

```python
M_x=np.float32([[1,0.2,0],
               [0.2,1,0],
               [0,0,1]])

sheared_img = cv2.warpPerspective(input_img,M_x,(cols,rows))

plt.axis('off')
plt.imshow(sheared_img)
plt.show()
```

```python
M_x=np.float32([[1,0.2,0],
               [0,1,0],
               [0,0,1]])

sheared_img_xaxis=cv2.warpPerspective(input_img,M_x,(cols,rows))

plt.axis('off')
plt.imshow(sheared_img_xaxis)
plt.show()
```

```python
M_x=np.float32([[1,0,0],
               [0.2,1,0],
               [0,0,1]])

sheared_img_xaxis=cv2.warpPerspective(input_img,M_x,(cols,rows))

plt.axis('off')
plt.imshow(sheared_img_xaxis)
plt.show()
```

# Image Reflection

```python
M_x=np.float32([[1,0,0],
               [0,-1,rows],
               [0,0,1]])

reflected_img_xaxis=cv2.warpPerspective(input_img,M_x,(cols,rows))

plt.axis('off')
plt.imshow(reflected_img_xaxis)
plt.show()
```

```python
M_x=np.float32([[-1,0,cols],
               [0,1,0],
               [0,0,1]])

reflected_img_yaxis=cv2.warpPerspective(input_img,M_x,(cols,rows))

plt.axis('off')
plt.imshow(reflected_img_xaxis)
plt.show()
```

# Image Rotation

```python
angle1=np.radians(5)
angle2=np.radians(35)
angle3=np.radians(5)
angle4=np.radians(45)
M=np.float32([[np.cos(angle1),-(np.sin(angle2)),0],
               [np.sin(angle3),np.cos(angle4),0],
               [0,0,1]])
rotated_img=cv2.warpPerspective(input_img,M,(cols,rows))

plt.axis('off')
plt.imshow(rotated_img)
plt.show()
```
